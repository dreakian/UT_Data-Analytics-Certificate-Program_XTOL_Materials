# What is the Problem?

### A company is trying to predict how much volume it should have for new products that it wants to start selling.

### There are two datasets. One is complete and the other incomplete. The primary problem is to try to predict the "volume" for each product.
### The complete dataset has each of the volumnes for all of its products.
### However, the incomplete dataset does not have any volume data for any of its products. This is because the incomplete dataset refers to products that the company has not started to sell yet.

# Understanding the data (analyzing the features, response variable, and thinking about data relevance for the business problem

### Both datasets include very different products, like printers, software, and accessories. Each product has information regarding its price, size, customer reception (ratings, recommendations), and volume. 

### The business is most interested in understanding how all of this information relates to the volume of any given product. That is, the business wants to know how much of a product it should have in inventory given a product's information.

### Given the results above, the following variables are highly correlated with Volume: 5 Star Reviews, 4 Star Reviews, 3 Star Reviews, Positive Service Reviews. Of particular note, 5 Star Reviews are entirely correlated with Volume, having a correlation value of 1. 

### The remaining variables are either moderately correlated with Volume (2 Star Reviews) or are negatively correlated. 

### Given these insights, I will create two feature spaces. 

### One feature space will include every feature, except "ProductNum", "BestSellerRank" and "ProfitMarigin". The first variable is simply an index and contains no insightful information. The second variable has a weak correlation with the target variable and attempts to compare products that aren't easily or meaningfully comparable. The third variable is also weakly correlated with the target variable and is not insightful for analysis.

### The second feature space will only include the variables that are moderately and highly correlated with Volume.
